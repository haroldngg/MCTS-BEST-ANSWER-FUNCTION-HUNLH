```python
## LES FONCTIONS DE HP
from deuces import Card, Evaluator, Deck
import itertools

evaluator = Evaluator()

def HandPotential_1(boardcards, ourcards):
    # Hand potential array, each index represents ahead, tied, and behind.
    HP = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]
    # Initialize HPTotal to 0.
    HPTotal = [0, 0, 0]

    deck = Deck.GetFullDeck()
    evaluator = Evaluator()

    ourrank = evaluator.evaluate(boardcards, ourcards)

    # Remove the cards from the deck.

    for card in boardcards + ourcards:
        deck.remove(card)

    for oppcards in itertools.combinations(deck, 2):

        oppcards = list(oppcards)

        # Remove the cards from the deck.
        for card in oppcards:
            deck.remove(card)

        opprank = evaluator.evaluate(boardcards, oppcards)
        if ourrank < opprank:
            index = 0 #ahead
        elif ourrank == opprank:
            index = 1 #tied
        else:
            index = 2 #behind

        # All possiblities of next card
        for next_card in deck:
            HPTotal[index] += 1
            updated_board = boardcards + [next_card]
            ourbest = evaluator.evaluate(updated_board, ourcards)
            oppbest = evaluator.evaluate(updated_board, oppcards)

            if ourbest < oppbest:
                HP[index][0] += 1
            elif ourbest == oppbest:
                HP[index][1] += 1
            else:
                HP[index][2] += 1

        # Restore the cards to the deck.
        for card in oppcards:
            deck.append(card)

    #print("HP: ", HP)
    #print("HPTotal: ", HPTotal)

    # if the last two row of HP is all 0, then Ppot = 0
    if HP[2][0] == 0 and HP[2][1] == 0 and HP[2][2] == 0 and HP[1][0] == 0 and HP[1][1] == 0 and HP[1][2] == 0:
        Ppot = 0
    else:
        Ppot = (HP[2][0] + HP[2][1]/2 + HP[1][0]/2) / (HPTotal[2] + HPTotal[1])

    Npot = (HP[0][2] + HP[1][2]/2 + HP[0][1]/2) / (HPTotal[0] + HPTotal[1])
    #print("Ppot: ", Ppot)
    #print("Npot: ", Npot)

    return [Ppot, Npot]


def HandPotential_2(boardcards, ourcards):
    # Hand potential array, each index represents ahead, tied, and behind.
    HP = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]
    # Initialize HPTotal to 0.
    HPTotal = [0, 0, 0]

    deck = Deck()

    ourrank = evaluator.evaluate(boardcards, ourcards)

    # Remove the cards from the deck.
    for card in boardcards + ourcards:
        deck.cards.remove(card)

    for oppcards in itertools.combinations(deck.cards, 2):
        oppcards = list(oppcards)

        # Remove the cards from the deck.
        for card in oppcards:
            deck.cards.remove(card)

        opprank = evaluator.evaluate(boardcards, oppcards)
        if ourrank < opprank:
            index = 0 #ahead
        elif ourrank == opprank:
            index = 1 #tied
        else:
            index = 2 #behind

        # All possiblities of the next 2 cards (turn and river)
        for next_cards in itertools.combinations(deck.cards, 2):
            next_cards = list(next_cards)
            HPTotal[index] += 1
            updated_board = boardcards + next_cards
            ourbest = evaluator.evaluate(updated_board, ourcards)
            oppbest = evaluator.evaluate(updated_board, oppcards)

            if ourbest < oppbest:
                HP[index][0] += 1
            elif ourbest == oppbest:
                HP[index][1] += 1
            else:
                HP[index][2] += 1

        # Restore the cards to the deck.
        for card in oppcards:
            deck.cards.append(card)

    print("HP: ", HP)
    print("HPTotal: ", HPTotal)

    # if the last two row of HP is all 0, then Ppot = 0
    if HP[2][0] == 0 and HP[2][1] == 0 and HP[2][2] == 0 and HP[1][0] == 0 and HP[1][1] == 0 and HP[1][2] == 0:
        Ppot = 0
    else:
        Ppot = (HP[2][0] + HP[2][1]/2 + HP[1][0]/2) / (HPTotal[2] + HPTotal[1])

    Npot = (HP[0][2] + HP[1][2]/2 + HP[0][1]/2) / (HPTotal[0] + HPTotal[1])
    print("Ppot: ", Ppot)
    print("Npot: ", Npot)

    return Ppot, Npot

if __name__ == "__main__":
    deck = Deck()
    hand = [Card.new('7d'), Card.new('5c')]
    board = [Card.new('4h'), Card.new('3c'), Card.new('2h')]

    HandPotential_1(board, hand)
    HandPotential_2(board, hand)
```

```python
##Les fonctions de HSE
import itertools
from deuces import Card, Evaluator, Deck
# from hse.hse_2 import hse_2
# from hse.hse_1 import hse_1





def hse_1(boardcards, ourcards):
    wins = 0
    losses = 0
    ties = 0
    deck = list(Deck.GetFullDeck())  # Create a copy of the deck
    evaluator = Evaluator()

    # Remove known cards from the deck
    known_cards = set(boardcards + ourcards)
    deck = [card for card in deck if card not in known_cards]

    hero_score = evaluator.evaluate(boardcards, ourcards)

    # Calculate against all possible opponent hands
    for opp_hand in itertools.combinations(deck, 2):
        villain_score = evaluator.evaluate(boardcards, list(opp_hand))

        if hero_score < villain_score:
            wins += 1
        elif hero_score > villain_score:
            losses += 1
        else:
            ties += 1

    hse = (wins + ties / 2) / (wins + losses + ties)
    return hse


def hse_2(board, hand):
    wins = 0
    losses = 0
    ties = 0

    deck = Deck()

    for card in board + hand:
        deck.cards.remove(card)

    hero_score = evaluator.evaluate(board, hand)
    hero_class = evaluator.get_rank_class(hero_score)

    # for all possible starting hands the opponent could have
    for oppcards_1 in itertools.combinations(deck.cards, 2):

        oppcards1 = list(oppcards_1)

        # Remove the cards from the deck.
        for card in oppcards1:
            deck.cards.remove(card)

        villain_score1 = evaluator.evaluate(board, oppcards1)

        # all possible hands of second opponent
        for oppcards_2 in itertools.combinations(deck.cards, 2):

            oppcards2 = list(oppcards_2)

            villain_score2 = evaluator.evaluate(board, oppcards2)

            if hero_score < min(villain_score1, villain_score2):
                wins += 1
            elif hero_score == min(villain_score1, villain_score2):
                ties += 1
            else:
                losses += 1

        # adding the cards back
        for card in oppcards1:
            deck.cards.append(card)

    print("Hero best hand:")
    print("Rank: ", hero_score)
    print("Class: ", evaluator.class_to_string(hero_class))

    print("Wins: ", wins)
    print("Losses: ", losses)
    print("Ties: ", ties)

    hse = (wins + ties / 2) / (wins + losses + ties)
    print("HSE_2: ", hse)
    return hse

if __name__ == "__main__":
    hand = [Card.new('2s'), Card.new('3c')]

    board = [Card.new('Th'), Card.new('8h'), Card.new('6h')]

    h_1 = hse_1(board, hand)
    print("------")
    h_2 = hse_2(board, hand)
    print("------")
    print("Adjusted HSE for 2 players:", pow(h_1, 2))
    print("HSE_2:", h_2)
```

```python
##ELIE
from math import sqrt, log, copysign
import random ## pour les mouvements aleatoires
from deuces import Card, Deck, Evaluator ##pour les cartes
from random import shuffle
import copy
import numpy as np
import time
from graphviz import Digraph
import matplotlib.pyplot as plt
import networkx as nx





def signe(x):
    if x== 0:
        return 0
    return copysign(1,x)

class Deck:
    """
    Class representing a deck. The first time we create, we seed the static
    deck with the list of unique card integers. Each object instantiated simply
    makes a copy of this object and shuffles it.
    """
    _FULL_DECK = []

    def __init__(self):

        self.shuffle()

    def shuffle(self):
        # and then shuffle
        self.cards = Deck.GetFullDeck()
        shuffle(self.cards)

    def draw(self, n=1):
        if n == 1:
            return self.cards.pop(0)

        cards = []
        for i in range(n):
            cards.append(self.draw())
        return cards

    def __str__(self):
        return Card.print_pretty_cards(self.cards)

    def copy(self):
        return self.cards.copy()


    @staticmethod
    def GetFullDeck():
        if Deck._FULL_DECK:
            return list(Deck._FULL_DECK)

        # create the standard 52 card deck
        for rank in Card.STR_RANKS:
            for suit,val in Card.CHAR_SUIT_TO_INT_SUIT.items():
                Deck._FULL_DECK.append(Card.new(rank + suit))

        return list(Deck._FULL_DECK)



class Node:
    """ A node in the game tree. Note wins is always from the viewpoint of playerJustMoved.
        Crashes if state not specified.
    """
    def __init__(self):##, move = None, parent = None, state :'PokerState' = None): # mettre current state?
        self.parentNode = None # "None" for the root node
        self.childNodes = [] ## les nodes enfants
        self.wins = 0
        self.visits = 0
        self.current='preflop'
        self.hand=[]
        self.flop_cards=[]
        self.turn_card=[]
        self.river_card=[]
        self.opp_hand=[]
        self.community_cards=[]
        self.preflop_moves=''
        self.flop_moves=''
        self.turn_moves=''
        self.river_moves=''
        self.opp_preflop_moves=''
        self.opp_flop_moves=''
        self.opp_turn_moves=''
        self.opp_river_moves=''
        self.deck = Deck.GetFullDeck()#Elie
        random.shuffle(self.deck)
        self.untried_moves = ["fold","check","raise"]
        self.reward=0
        self.pot=10
        self.my_pot=10
        self.absolute_win=0
        #if pr remplir les cartes





    def hand_strength_potential(hand, board):
      h = [Card.new(carte) for carte in hand]
      b = [Card.new(carte) for carte in board]
      print(type(h[0]))
      print(h[0])
      if len(b)<5:
          try:
              hpp, hpn = HandPotential_1(b, h)
          except:
              print(board, hand)
      else:
          hpp, hpn = 0,0
      return (hse_1(b,h),hpp,hpn)
    def card_value(card):
        s=Card.int_to_pretty_str(card)[0]
        if s=='T':
            return 10
        elif s=='J':
            return 11
        elif s=='Q':
            return 12
        elif s=='K':
            return 13
        elif s=='A':
            return 14
        else:
            return int(s)
    def card_color(card):
      s=Card.int_to_pretty_str(card)[1]
      if s=='❤':
          return 1
      elif s=='♦':
          return 2
      elif s=='♣':
          return 3
      else:
          return 4

    def opp_preflop_move(self):
        stack = 100
        flop_position = 1  # heads_up et agit toujours en second donc on le considère en big-blind
        nb_joueur_au_depart = 2
        pot_size_before_decision = self.pot  # a prendre probablement comme argument de la fonction play
        nb_unfolded_players = 0
        money_to_add_to_call = 10 ## Il faudrait rajouter un parametre car ca pourrait etre 0 ou 10 selon la situation
        #hand_strength = hse_1(self.community_cards,self.hand)
        #hand_neg_pot = HandPotential_1(self.community_cards,self.hand)[1]
        #hand_pos_pot = HandPotential_1(self.community_cards,self.hand)[0]
        last_raise_preflop_player = 0 ## il faudrait rajouter un parametre
        #last_raise_flop_player = 0 ##il faudrait rajouter un parametre
        #mise_en_cas_raise = 10
        top_card = max(card_value(self.opp_hand[0]), card_value(self.opp_hand[1]))
        bottom_card=min(card_value(self.opp_hand[0]), card_value(self.opp_hand[1]))
        diff_between_cards = abs(card_value(self.opp_hand[0]) - card_value(self.opp_hand[1]))
        diff_2_highest = max(card_value(self.opp_hand[0]), card_value(self.opp_hand[1])) - min(card_value(self.opp_hand[0]), card_value(self.opp_hand[1]))
        if card_color(self.opp_hand[0]) == card_color(self.opp_hand[1]):
            color = 1
        else:
            color = 0
        #hand_strength, hand_pos_pot, hand_neg_pot = hand_strength_potential(self.opp_hand, self.community_cards)
        pre = np.array([stack, flop_position, nb_joueur_au_depart, pot_size_before_decision, nb_unfolded_players, money_to_add_to_call, last_raise_preflop_player, top_card, bottom_card, color]).reshape(1, -1)
        predict = model_preflop.predict(pre,verbose=0)
        if predict[0][0] > 0.5:
            return 'raise'
        elif predict[0][1] > 0.5:
            return 'call'
        else:
            return 'fold'

    def opp_flop_move(self):
        stack = 100
        flop_position = 1  # heads_up et agit toujours en second donc on le considère en big-blind
        nb_joueur_au_depart = 2
        pot_size_before_decision = self.pot  # a prendre probablement comme argument de la fonction play
        nb_unfolded_players = 0
        money_to_add_to_call = 10
        hand_strength = hse_1(self.community_cards,self.opp_hand)
        hand_neg_pot = HandPotential_1(self.community_cards,self.opp_hand)[1]
        hand_pos_pot = HandPotential_1(self.community_cards,self.opp_hand)[0]
        last_raise_preflop_player = 0
        last_raise_flop_player = 0
        #mise_en_cas_raise = 10
        top_card = max(card_value(self.opp_hand[0]), card_value(self.opp_hand[1]))
        diff_between_cards = abs(card_value(self.opp_hand[0]) - card_value(self.opp_hand[1]))
        compcard=[card_value(self.community_cards[0]), card_value(self.community_cards[1]), card_value(self.community_cards[2])]
        compcard.sort()
        diff_2_highest = compcard[2]-compcard[1]
        diff_1_2=compcard[1]-compcard[0]
        top_card_board=max([card_value(self.community_cards[0]), card_value(self.community_cards[1]), card_value(self.community_cards[2])])
        if card_color(self.opp_hand[0]) == card_color(self.opp_hand[1]):
            color = 1
        else:
            color = 0
        #hand_strength, hand_pos_pot, hand_neg_pot = hand_strength_potential(self.opp_hand, self.community_cards)
        pre = np.array([stack, flop_position, nb_joueur_au_depart, pot_size_before_decision, nb_unfolded_players, money_to_add_to_call, hand_strength, hand_pos_pot, hand_neg_pot, last_raise_preflop_player,last_raise_flop_player, top_card, diff_between_cards, diff_2_highest,diff_1_2,top_card_board, color]).reshape(1, -1)
        predict = model_flop.predict(pre,verbose=0)
        if predict[0][0] > 0.5:
            return 'raise'
        elif predict[0][1] > 0.5:
            return 'call'
        else:
            return 'fold'


    def opp_river_move(self):
        stack = 100
        river_position = 1  # heads_up et agit toujours en second donc on le considère en big-blind
        nb_joueur_au_depart = 2
        pot_size_before_decision = self.pot  # a prendre probablement comme argument de la fonction play
        nb_unfolded_players = 0
        money_to_add_to_call = 10 ## a revoir
        hand_strength = hse_1(self.community_cards,self.hand)
        #hand_neg_pot = HandPotential_1(self.community_cards,self.hand)[1]
        #hand_pos_pot = HandPotential_1(self.community_cards,self.hand)[0]
        last_raise_turn_player = 0
        last_raise_river_player = 0
        #mise_en_cas_raise = 10
        top_card_player = max(card_value(self.opp_hand[0]), card_value(self.opp_hand[1]))
        diff_between_cards_player = abs(card_value(self.opp_hand[0]) - card_value(self.opp_hand[1]))
        compcard=[card_value(i) for i in self.community_cards]
        compcard.sort()
        diff_2_3_highest = compcard[2]-compcard[1]
        diff_1_2_highest=compcard[1]-compcard[0]
        diff_3_4_highest=compcard[3]-compcard[2]
        diff_4_5_highest=compcard[4]-compcard[3]
        top_card_board=compcard[4]
        if card_color(self.opp_hand[0]) == card_color(self.opp_hand[1]):
            color = 1
        else:
            color = 0
        #hand_strength, hand_pos_pot, hand_neg_pot = hand_strength_potential(self.opp_hand, self.community_cards)
        pre = np.array([stack, river_position, nb_joueur_au_depart, pot_size_before_decision, nb_unfolded_players, money_to_add_to_call, hand_strength, last_raise_turn_player,last_raise_river_player, top_card_player,top_card_board, diff_between_cards_player, diff_1_2_highest,diff_2_3_highest,diff_3_4_highest,diff_4_5_highest, color]).reshape(1, -1)
        predict = model_flop.predict(pre,verbose=0)
        if predict[0][0] > 0.5:
            return 'raise'
        elif predict[0][1] > 0.5:
            return 'call'
        else:
            return 'fold'

    def opp_turn_move(self):
        stack = 100
        turn_position = 1  # heads_up et agit toujours en second donc on le considère en big-blind
        nb_joueur_au_depart = 2
        pot_size_before_decision = self.pot  # a prendre probablement comme argument de la fonction play
        nb_unfolded_players = 0
        money_to_add_to_call = 10 ## a revoir
        hand_strength = hse_1(self.community_cards,self.hand)
        hand_neg_pot = HandPotential_1(self.community_cards,self.hand)[1]
        hand_pos_pot = HandPotential_1(self.community_cards,self.hand)[0]
        last_raise_flop_player = 0
        last_raise_turn_player = 0
        #mise_en_cas_raise = 10
        top_card_player = max(card_value(self.opp_hand[0]), card_value(self.opp_hand[1]))
        diff_between_cards_player = abs(card_value(self.opp_hand[0]) - card_value(self.opp_hand[1]))
        compcard=[card_value(i) for i in self.community_cards]
        compcard.sort()
        diff_2_3_highest = compcard[2]-compcard[1]
        diff_1_2_highest=compcard[1]-compcard[0]
        diff_3_4_highest=compcard[3]-compcard[2]
        top_card_board=compcard[3]
        if card_color(self.opp_hand[0]) == card_color(self.opp_hand[1]):
            color = 1
        else:
            color = 0
        #hand_strength, hand_pos_pot, hand_neg_pot = hand_strength_potential(self.opp_hand, self.community_cards)
        #attention on aurait un paramètre en moins a voir
        pre = np.array([stack, turn_position, nb_joueur_au_depart, pot_size_before_decision, nb_unfolded_players, money_to_add_to_call, hand_strength, hand_pos_pot, hand_neg_pot, last_raise_flop_player, top_card_player, diff_between_cards_player,diff_2_3_highest, diff_1_2_highest,diff_3_4_highest,top_card_board, color]).reshape(1, -1)
        predict = model_flop.predict(pre,verbose=0)
        if predict[0][0] > 0.5:
            return 'raise'
        elif predict[0][1] > 0.5:
            return 'call'
        else:
            return 'fold'
    def UCTSelectChild(self):
            """ Use the UCB1 formula to select a child node. Often a constant UCTK is applied so we have
                lambda c: c.wins/c.visits + UCTK * sqrt(2*log(self.visits)/c.visits to vary the amount of
                exploration versus exploitation.
            """
            if(random.random()<0.1):
              s=random.choice(self.childNodes)
            else:
              s = sorted(self.childNodes, key = lambda c: c.wins/c.visits + sqrt(2*log(self.visits)/c.visits))[-1]
            return s

    def clone(self):#effet de bord?
        # Création d'une nouvelle instance de la classe Node
        cloned_node = Node()  # state est None car playerJustMoved est déjà copié
        # Copie des valeurs des attributs
        #parent?
        cloned_node.childNodes = self.childNodes[:]  # Copie superficielle de la liste
        cloned_node.wins = self.wins
        cloned_node.visits = self.visits
        cloned_node.current = self.current
        cloned_node.hand = self.hand[:]
        cloned_node.flop_cards = self.flop_cards[:]
        cloned_node.turn_card = self.turn_card
        cloned_node.river_card = self.river_card
        cloned_node.opp_hand = self.opp_hand[:]
        cloned_node.community_cards = self.community_cards[:]
        cloned_node.preflop_moves = self.preflop_moves[:]
        cloned_node.flop_moves = self.flop_moves
        cloned_node.turn_moves = self.turn_moves
        cloned_node.river_moves = self.river_moves
        cloned_node.opp_preflop_moves = self.opp_preflop_moves
        cloned_node.opp_flop_moves = self.opp_flop_moves
        cloned_node.opp_turn_moves = self.opp_turn_moves
        cloned_node.opp_river_moves = self.opp_river_moves
        cloned_node.deck = self.deck
        cloned_node.untried_moves = self.untried_moves
        cloned_node.reward = self.reward
        cloned_node.pot=self.pot
        cloned_node.my_pot = self.my_pot
        return cloned_node

    def make_graph(self, graph=None):
        if graph is None:
            graph = Digraph(comment='Poker Game Tree')
            graph.attr('node', shape='ellipse')

        # Ajouter le noeud courant au graphique
        graph.node(name=str(id(self)),
                  label=f"W/V: {self.wins}/{self.visits}\nMoves: {self.preflop_moves}, {self.flop_moves}, {self.turn_moves}, {self.river_moves}")

        # Ajouter des arêtes et des noeuds enfants récursivement
        for child in self.childNodes:
            graph.edge(str(id(self)), str(id(child)))
            child.make_graph(graph)

        return graph

    def AddChild(self, s):##ici il me faut une maniere differentes d'ajouter un enfant
        """ Remove m from untriedMoves and add a new child node for this move.
            Return the added child node
            Elie speaking:
            ##Idee: creer tte les nodes avec comme enfant (check, raise, fold) des le debut
            # on les mets dans Untried Move et apres on tire au hasard dans untried move en priorité?
        """
        ##la fonction initial def AddChild(self, m, s):##ici il me faut une maniere differentes d'ajouter un enfant
        ##il faudrait verifier qu'il n'a pas encore d'enfant avec les memes cartes
        for i in self.childNodes: ##on verfie qu'il n'existe pas deja des enfants contenant ces cartes et on traitera cela
            if(i.preflop_moves == s.preflop_moves and i.flop_moves == s.flop_moves and i.turn_moves == s.turn_moves and i.river_moves == s.river_moves):
                return
        self.childNodes.append(s)
        return n
    def IsAChild(self, s):##cette fonction permet de verifier si un noeud est deja un enfant # Elie: pq pas verifier avec self.parent
        for i in self.childNodes: ##on verfie qu'il n'existe pas deja des enfants contenant ces cartes et on traitera cela
            if(i.hand==s.hand): ## dans l'idee il faudrait comparer tous les etats du jeu
                return True
        return False

    ##def Update(self, result): #Elie: ou est le pb?
      ##  __init__(self, move = None, parent = None, state :'PokerState' = None)
        """ Update this node - one additional visit and result additional wins. result must be from the viewpoint of playerJustmoved.
        """
        ##self.visits += 1
        ##self.wins += result

    def find_move(self):
        if (self.river_moves !=''):
            return self.river_moves
        if (self.turn_moves !=''):
            return self.turn_moves
        if (self.flop_moves !=''):
            return self.flop_moves
        if (self.preflop_moves !=''):
            return self.preflop_moves

    def __repr__(self):
        return "[ W/V:" + str(self.wins) + "/" + str(self.visits)  + "]"

    def result(self):
        return np.array([self.wins,self.visits])

    def TreeToString(self, indent):
        s = self.IndentString(indent) + str(self)
        for c in self.childNodes:
             s += c.TreeToString(indent+1)
        return s

    def IndentString(self,indent):
        s = "\n"
        for i in range (1,indent+1):
            s += "| "
        return s

    def ChildrenToString(self):
        s = ""
        for c in self.childNodes:
             s += str(c) + "\n"
        return s

    def backpropagate(self, reward):#Elie: on mets -1 pr la perte? et 0 pr egalité?
        """
        Update the win and visit count for the node. Propagate the update up to the root node.

        Args:
            is_win (bool): True if the current path resulted in a win, otherwise False.
        """
        self.wins += reward
        self.visits += 1
        #self.pot=0

        if self.parentNode != None:
            self.parentNode.backpropagate(reward)
        else:#pr reinitialiser un pot
          if self.current=='preflop':
            self.pot=10
          elif self.current =='flop':
            self.pot = 20
          elif self.current == 'turn':
            self.pot = 30
          elif self.current =='river':
            self.pot = 40


    def play(self): ## ici le but est d'entrainer notre arbre plusieurs fois pour qu'il puisse apprendre a jouer l'idee c'est qu'a chque modification on cree un enfant donc il faut etre meticuleux
        #print(self.current)
        #print(Card.print_pretty_cards(self.hand))
        #print(self.current)
        #if (self.opp_hand):
            #Card.print_pretty_cards(self.opp_hand)

        #Card.print_pretty_cards(self.community_cards)

        if self.current=='preflop' :#avant de recevoir les cartes
            if self.hand==[]: ## la main est vide donc on ajoute les cartes du joueurs nous sommes donc dans la root
                self.hand.append(self.deck.pop())
                self.hand.append(self.deck.pop())
                self.opp_hand.append(Card.new('Ah'))
                self.opp_hand.append(Card.new('Ts'))




                # pas d'enfant on en cree un, mettre if untried moves pas vide apres
                # on a recu les cartes, et on tire ici les 3 du flop


            deck = copy.deepcopy(self.deck)
            random.shuffle(deck)
            #deck.shuffle()
            card_opp1 = deck.pop()
            card_opp2 = deck.pop()
            #self.opp_hand[0]=deck.pop() #ca compte pas
            #self.opp_hand[1] = deck.pop()

            card1 = deck.pop()
            card2 = deck.pop()
            #print(Card.print_pretty_card(card1))
            card3 = deck.pop()
            flop_cards = [card1, card2, card3]
            community_cards = [card1, card2, card3]


            if (self.untried_moves):
                my_choice = self.untried_moves.pop(random.randrange(0,len(self.untried_moves)))#take a random choice and pop it from untried_moves

                if (my_choice == 'dead end'):
                    self.untried_moves.append(my_choice) # On remets dead end psk c'est une feuille
                    self.backpropagate(-1*self.my_pot)

                elif (my_choice == 'fold'):
                    s=self.clone()
                    s.parentNode= self
                    s.flop_cards= flop_cards
                    s.community_cards = community_cards
                    s.opp_hand[0]= card_opp1
                    s.opp_hand[1] = card_opp2
                    print(self.opp_hand[0])

                    s.childNodes = [] ## les nodes enfants
                    s.wins = 0
                    s.visits = 1
                    s.current='flop'
                    s.preflop_moves='fold'

                    s.deck = deck#Elie
                    s.untried_moves = ['dead end']#ends here

                    self.AddChild(s)
                    s.backpropagate(-1*s.my_pot)
                    #self.play()

                else:#hypothese simplificatrice, soit nous suit soit fold, ie call peut etre bet en cas de raise
                    b=False
                    if(my_choice=='raise'):
                      self.pot+=10
                      self.my_pot +=10
                      b = True
                    opp_choice= random.choice(['call','fold'])
                    self.opp_preflop_moves=opp_choice#le premier move ici

                    s=self.clone()
                    s.parentNode= self
                    s.childNodes=[]
                    s.wins=0
                    s.visits=1
                    s.current='flop'
                    s.flop_cards= flop_cards
                    s.opp_hand[0]= card_opp1
                    s.opp_hand[1] = card_opp2

                    s.community_cards = community_cards
                    s.deck = deck#Elie

                    s.untried_moves = ["fold","check","raise"]
                    s.preflop_moves= my_choice
                    s.opp_preflop_moves= opp_choice

                    self.AddChild(s)
                    if(opp_choice=='fold'):
                        self.backpropagate(self.pot-self.my_pot)
                    else:
                        temp = 10*b
                        s.pot+= temp #si le opps a call alors qu'on a raise, le pot augmente
                        s.play()
                        s.pot -= temp
            else:
                #if child Nodes n'est pas vide
                s = self.UCTSelectChild()
                #s.visits = 0
                #if(s.preflop_moves=='raise'):
                  #s.pot+=10
                s.flop_cards=flop_cards
                s.community_cards=community_cards
                s.opp_hand[0]= card_opp1
                s.opp_hand[1] = card_opp2
                #determiner s.opp_moves
                #if fold backpropagate 1
                #else s.play()

                s.deck = deck

                s.play()

        elif self.current=='flop':#il y a deja le flop, il faut tirer la turn pr la suite
            deck = copy.deepcopy(self.deck)
            random.shuffle(deck)

            turn_card= deck.pop()

            if (self.untried_moves):
                my_choice = self.untried_moves.pop(random.randrange(0,len(self.untried_moves)))#take a random choice and pop it from untried_moves


                if (my_choice == 'dead end'):
                    self.untried_moves.append(my_choice) # On remets dead end psk c'est une feuille
                    self.backpropagate(-1*self.my_pot)

                elif (my_choice=='fold'):
                    s=self.clone()
                    s.parentNode = self # "None" for the root node
                    s.childNodes = [] ## les nodes enfants
                    s.wins = 0
                    s.visits = 1
                    s.current='turn'
                    s.turn_card = turn_card

                    s.community_cards = s.community_cards[0:3]
                    s.community_cards.append(turn_card)

                    s.flop_moves=my_choice

                    s.deck = deck#Elie
                    s.untried_moves = ['dead end']#ends here
                    s.backpropagate(-1*self.my_pot)
                    self.AddChild(s)
                else:#hyp simplificatrice encore
                    opp_choice= random.choice(['call','fold'])
                    opp_choice=self.opp_flop_move()
                    b=False

                    self.opp_flop_moves =opp_choice

                    s=self.clone()
                    if(opp_choice=='raise'):
                      opp_choice='call'
                      b=True
                      s.pot+=10
                    s.parentNode= self
                    s.childNodes=[]
                    s.wins=0
                    s.visits=1
                    s.current='turn'
                    s.turn_card=turn_card
                    s.community_cards = s.community_cards[0:3]
                    s.community_cards.append(turn_card)
                    s.flop_moves=my_choice
                    s.opp_flop_moves=opp_choice
                    if(my_choice=='raise'):
                      s.pot+=10 +10*b
                      s.my_pot +=10+10*b #si il avit deja raise on rentre puis on reraise
                    s.deck = deck
                    s.untried_moves = ["fold","check","raise"]
                    self.AddChild(s)
                    if(opp_choice=='fold'):
                        s.backpropagate(s.pot-s.my_pot)
                        #self.backpropagate(self.pot)
                    else:
                        s.pot +=10

                        s.play()
                        s.pot -=10*b #pr le premier raise du opps
                        s.pot -=10*b #pr le notre
                        s.my_pot-=10*b
                        s.pot -=10 # pr celui 5 lignes au dessus
            else:
                #if child Nodes n'est pas vide
                s = self.UCTSelectChild()
                #if(s.flop_moves=='raise'):
                  #s.pot+=10
                s.community_cards = self.community_cards[0:3]
                s.community_cards.append(turn_card)
                s.opp_hand = self.opp_hand
                s.turn_card = turn_card
                #if(len(s.community_cards)!=3):
                #print("AKHAA")
                #Card.print_pretty_cards(s.community_cards)


                s.deck = deck
                s.play()



        elif self.current=='turn':
            deck = copy.deepcopy(self.deck)
            random.shuffle(deck)
            river_card= deck.pop()

            if (self.untried_moves):
                my_choice = self.untried_moves.pop(random.randrange(0,len(self.untried_moves)))#take a random choice and pop it from untried_moves

                if (my_choice == 'dead end'):
                    self.untried_moves.append(my_choice) # On remets dead end psk c'est une feuille
                    self.backpropagate(-1*self.pot)

                elif (my_choice=='fold'):
                    s=self.clone()
                    s.parentNode = self # "None" for the root node
                    s.childNodes = [] ## les nodes enfants
                    s.wins = 0
                    s.visits = 1
                    s.current='river'
                    s.river_card = river_card
                    if len(s.community_cards) == 4:
                      s.community_cards.append(river_card)
                    else:
                      s.community_cards[4]= river_card

                    s.turn_moves=my_choice

                    s.deck = deck#Elie
                    s.untried_moves = ['dead end']#ends here
                    s.backpropagate(-1*self.pot)
                    self.AddChild(s)
                else:#hyp simplificatrice encore
                    opp_choice= random.choice(['call','fold'])
                    opp_choice=self.opp_turn_move()
                    b=False

                    self.opp_flop_moves =opp_choice

                    #on cree l'enfant et on continue
                    s=self.clone()
                    if(opp_choice=='raise'):
                      opp_choice='call'
                      b=True
                      s.pot+=10
                    s.parentNode= self
                    s.childNodes=[]
                    s.wins=0
                    s.visits=1
                    s.current='river'
                    s.river_card = river_card
                    if len(s.community_cards) == 4:
                      s.community_cards.append(river_card)
                    else:
                      s.community_cards[4]= river_card
                    s.turn_moves=my_choice
                    s.opp_turn_moves=opp_choice
                    s.deck = deck
                    if(my_choice=='raise'):
                      s.pot+=10 +10*b
                      s.my_pot +=10+10*b
                    s.untried_moves = ["fold","check","raise"]
                    self.AddChild(s)
                    if(opp_choice=='fold'):
                        s.backpropagate(s.pot)
                        #self.backpropagate(self.pot)
                    else:
                        s.pot +=10
                        s.play()
                        s.pot -=10*b #pr le premier raise du opps
                        s.pot -=10*b #pr le notre
                        s.my_pot-=10*b
                        s.pot -=10
            else:
                #if child Nodes n'est pas vide
                s = self.UCTSelectChild()
                #if(s.turn_moves=='raise'):
                  #s.pot+=10
                s.community_cards = self.community_cards
                s.opp_hand = self.opp_hand
                #if(len(s.community_cards)!=4):
                    #print("AKHAAAAAAAA")
                #Card.print_pretty_cards(s.community_cards)
                s.river_card = river_card
                if len(s.community_cards) == 4:
                    s.community_cards.append(river_card)
                else:
                    s.community_cards[4]= river_card
                s.deck = deck
                s.play()










        elif self.current=='river':#la river est tirée, il faut faire un move
            if (self.untried_moves):
                my_choice = self.untried_moves.pop(random.randrange(0,len(self.untried_moves)))#take a random choice and pop it from untried_moves


                if (my_choice == 'dead end'):
                    self.untried_moves.append(my_choice) # On remets dead end psk c'est une feuille
                    self.backpropagate(-1*self.pot)

                elif (my_choice=='fold'):
                    s=self.clone()
                    s.parentNode = self # "None" for the root node
                    s.childNodes = [] ## les nodes enfants
                    s.wins = 0
                    s.visits = 1
                    s.current='showdown'

                    s.river_moves=my_choice

                    s.untried_moves = ['dead end']#ends here
                    s.backpropagate(-1*self.pot)
                    self.AddChild(s)
                else:#hyp simplificatrice encore
                    opp_choice= random.choice(['call','fold'])
                    opp_choice=self.opp_river_move()
                    b=False

                    self.opp_flop_moves =opp_choice

                    #on cree l'enfant et on continue
                    s=self.clone()
                    if(opp_choice=='raise'):
                      opp_choice='call'
                      b=True
                      s.pot+=10
                    s.parentNode= self
                    s.childNodes=[]
                    s.wins=0
                    s.visits=1
                    s.current='showdown'
                    if(my_choice=='raise'):
                      s.pot+=10 +10*b
                      s.my_pot +=10+10*b #+10*b #10 de plus qu'en dessous
                    #if (my_choice == 'check'):
                      #s.pot +=10*b #s'il a raise et on a call, le pot augmente
                    #on peut pas augmenter le pot tt le temps


                    s.river_moves=my_choice
                    s.opp_river_moves=opp_choice

                    s.untried_moves = []#nothing else for now
                    self.AddChild(s)
                    if(opp_choice=='fold'):
                        s.backpropagate(s.pot)
                        #self.backpropagate(self.pot) ########A changer pour voir
                    else:

                        s.pot +=10
                        s.play()
                        s.pot -=10*b #pr le premier raise du opps
                        s.pot -=10*b #pr le notre
                        s.my_pot-=10*b
                        s.pot -=10
            else:
                #if child Nodes n'est pas vide
                s = self.UCTSelectChild()
                #if(s.river_moves=='raise'):
                  #s.pot+=10
                s.community_cards = self.community_cards
                s.opp_hand = self.opp_hand


                s.play()






        elif self.current=='showdown':



            evaluator = Evaluator()
            #tirer des cartes pr le opps
            #self.opp_hand.append(self.deck.pop())
            #self.opp_hand.append(self.deck.pop())
            #print(len(self.community_cards))
            #print(len(self.deck))
            #print(len(self.opp_hand))
            #Card.print_pretty_cards(self.hand)
            Card.print_pretty_cards(self.opp_hand)
            #Card.print_pretty_cards(self.community_cards)
            reward = evaluator.evaluate(self.hand,self.community_cards) - evaluator.evaluate(self.opp_hand,self.community_cards)
            #print(f'la reward est {reward}')
            #print(f'Le signe de reward est {signe(reward)}')
            if(signe(reward)==0):
              reward=float(self.pot)/2
            else:
              reward=signe(reward)*self.pot
            self.backpropagate(reward)
            self.reward=reward
            self.current='end'
        elif self.current == 'end':
            self.backpropagate(self.reward)

            #print('JL a tort')


def add_nodes_edges(G, node, pos, x_increment=32, y_increment=1, current_depth=0):
    y = -current_depth
    pos[node] = (pos[node][0], y)
    current_depth += 1
    dico = {'fold':0, 'check': -1 , 'raise' : 1}
    for child in node.childNodes:

        pos[child] = (pos[node][0] + dico.get(child.find_move())*x_increment, y - y_increment)
        G.add_edge(node, child)
        add_nodes_edges(G, child, pos, x_increment/2, y_increment, current_depth)




# Function to draw the tree with wins and visits information
def draw_tree(root):
    G = nx.DiGraph()
    G.add_node(root)
    pos = {root: (0, 0)}
    add_nodes_edges(G, root, pos)
    wins = {node: f'{node.find_move()}\n  Wins: {node.wins}\nVisits: {node.visits}' for node in G.nodes()}
    nx.draw(G, pos, labels=wins, with_labels=True, node_size=500, node_color='skyblue', font_size=6)
    plt.show()




```
